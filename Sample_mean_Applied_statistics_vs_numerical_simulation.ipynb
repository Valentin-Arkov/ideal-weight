# Applied statistics. Confidence band for sample mean

## Sample mean vs population mean

$x \sim N(\mu, \sigma) $

$\overline{x} \sim N(\mu, \frac{\sigma}{\sqrt n}) $

with n > 100:

Student Distribution --> Gaussian Distribution

## Example

$x \sim N(0, 10) $

$n = 1000$

$\overline{x} \sim N(0, \frac{10}{\sqrt {1000}}) $

```python
10 / 1000 ** 0.5
```

$\overline{x} \sim N(0, 0.32) $

## "Three sigma" rule

$ -1 \leq \overline{x} \leq +1 $

$p = 0.997$

# Numerical simulation experiment

## Generate sample data

```python
# Library for numeric arrays
import numpy as np
```

```python
# Initialize random number generator
np.random.seed(12345)
```

```python
# Sample size
n = 1000
```

```python
x = np.random.normal(loc=0, scale=10, size=n)
```

```python
# Sample mean and standard deviation (STD)
# loc=0, scale=10
np.mean(x), np.std(x)
```

```python
# Graphics module
import  matplotlib.pyplot as plt
```

```python
# Distribution histogram with # of bins
# Bias-variance tradeoff
plt.hist(x, bins=25)
plt.show()
```

```python
# Sample mean is an estimate of population mean
sample_mean = np.mean(x)
sample_mean
```

## Series of simulations

```python
sample_means = []
for i in range(10):
  x = np.random.normal(loc=0, scale=10, size=n)
  sample_means += [np.mean(x)]
```

# Comparison

```python
# Distribution of sample mean
plt.hist(sample_means, bins=20, edgecolor='black', color='white', label='Histogram of sample mean')
plt.xlabel("$\overline{x}$")
plt.legend()
plt.show()
```

```python
# Probability density: histogram of sample mean
plt.hist(sample_means, bins=20, density=True, edgecolor='black', color='white', label='Histogram of sample mean')
plt.xlabel("$\overline{x}$")
plt.legend()
plt.show()
```

```python
from scipy.stats import norm

# Mean and standard deviation for sample mean
sample_mean_mean = 0
sample_mean_std = 10 / 1000 ** 0.5

# Array of x-values to plot
xx = np.linspace(-1, +1, 100)

# Calculate PDF values
pdf = norm.pdf(xx, sample_mean_mean, sample_mean_std)

# Plot the PDF
plt.plot(xx, pdf, color='blue', linewidth=2)
plt.ylabel('Probability Density')
plt.title('Gaussian PDF')
plt.show()
```

```python
# Joint plot

plt.hist(sample_means, bins=20, density=True, edgecolor='black', color='white', label='Histogram of sample mean')
plt.plot(xx, pdf, color='blue', linewidth=2, label="Gaussian PDF")
plt.title('Sample mean distribution')
plt.xlabel("$\overline{x}$")
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
